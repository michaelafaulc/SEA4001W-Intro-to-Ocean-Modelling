```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read CSV Data
data = pd.read_csv('C:/Users/faulc/OneDrive - University of Cape Town/Desktop/michaela/ocean_modelling/activity1/flynn94.csv')

# Step 2: Plotting
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(data['time'], data['cells'])  # Assuming 'time' and 'cells' are column names in your CSV file
plt.title('Phytoplankton growth over time')
plt.xlabel('Time (days)')
plt.ylabel('Cells ($10^6$ ml$^{-1}$)')
plt.grid(True)
plt.show()
```

## Questions

**1) Is the growth exponential?**

Yes.

**2) Can you separate the growth curve into some phases?**

Yes.The growth curve can be divided into three phases. The first phase has no growth from day 0 to 5 as there are very few phytoplankton thus little growth. In the exponential phase from around day 6 to 18 there are enough resources and more phytoplankton thus there is rapid growth. This bloom is followed by the plateau from day 20 to 25 where the resources become depleted and no more cells can be supported.
