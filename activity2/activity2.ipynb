```python
import numpy as np
import matplotlib.pyplot as plt

def p_t(p_0, n_0, k_prime, t):
    """
    Calculate P(t) based on the given initial conditions and time.
    """
    return p_0 * ((p_0 + n_0) / (p_0 + n_0 * np.exp(-k_prime * (p_0 + n_0) * t)))

def n_t(p_0, n_0, k_prime, t):
    """
    Calculate N(t) based on the given initial conditions and time.
    """
    p_t_value = p_t(p_0, n_0, k_prime, t)
    return p_0 + n_0 - p_t_value

# Given initial conditions and parameters
p_0 = 0.1
n_0 = 5
k_prime = 0.1
time = np.arange(0, 21, 1)

# Calculate P(t) and N(t)
p_values = p_t(p_0, n_0, k_prime, time)
n_values = n_t(p_0, n_0, k_prime, time)

# Plot P(t) and N(t)
plt.figure(figsize=(10, 6))
plt.plot(time, p_values, label='P(t)', color='b')
plt.plot(time, n_values, label='N(t)', linestyle='--', color='r')
plt.title('Phytoplankton and Nutrient Concentrations Over Time')
plt.xlabel('Time (d)')
plt.ylabel('Concentration ($mmol$ $N$ $m^{-3}$) ')
plt.grid(True)
plt.legend()
plt.show()
```

**1) Why are the curves identical in shape?**

The curves are identical in shape because they share the same paramters and the same decay constant. The rate at which the phytoplankton increases is the same as the rate that the nutrient decreases. The matter is conserved and simply transferred from one form to the other.

When we increase the percentage of food consumed per day, the growth curve for phytoplankton and the decay curve for the nutrient both reach the point where there are no more nutrients available sooner compared to if the k' value was smaller. The graphs are steeper as the k' determines the steepness of the curves.

**3) What happens if the initial biomass of phytoplankton is close to the amount of resource available? How does the shape of the curve change?**


```python
def p_t(p_0, n_0, k_prime, t):
    """
    Calculate P(t) based on the given initial conditions and time.
    """
    return p_0 * ((p_0 + n_0) / (p_0 + n_0 * np.exp(-k_prime * (p_0 + n_0) * t)))

def n_t(p_0, n_0, k_prime, t):
    """
    Calculate N(t) based on the given initial conditions and time.
    """
    p_t_value = p_t(p_0, n_0, k_prime, t)
    return p_0 + n_0 - p_t_value

# Given initial conditions and parameters
p_0 = 3
n_0 = 5
k_prime = 0.1
time = np.arange(0, 21, 1)

# Calculate P(t) and N(t)
p_values = p_t(p_0, n_0, k_prime, time)
n_values = n_t(p_0, n_0, k_prime, time)

# Plot P(t) and N(t)
plt.figure(figsize=(10, 6))
plt.plot(time, p_values, label='P(t)', color='b')
plt.plot(time, n_values, label='N(t)', linestyle='--', color='r')
plt.title('Phytoplankton and Nutrient Concentrations Over Time')
plt.xlabel('Time (d)')
plt.ylabel('Concentration ($mmol$ $N$ $m^{-3}$) ')
plt.grid(True)
plt.legend()
plt.show()
```

When there are more phytoplankton in the initial state and this concentration is close to the concentration of available nutrients, then the point where the concentrations equal one another occurs sooner in time and the growth and decay curves no longer have an exponential shape. There is no longer an initial period of stable, slow growth or decay. The steepness of the curves is not directly influenced as k' is not changed.
